TASK4

```python
# Existing data
products = ["Laptop", "Mouse", "Keyboard", "Monitor", "Charger", "Headphones"]

categories = ["Electronics", "Accessories", "Accessories", "Electronics", "Accessories", "Accessories"]

price_dict = {
    "Laptop": 55000,
    "Mouse": 600,
    "Keyboard": 1200,
    "Monitor": 9000,
    "Charger": 800,
    "Headphones": 2500
}

# 1. Create catalog as list of tuples
catalog = []

for i in range(len(products)):
    name = products[i]
    price = price_dict[name]
    category = categories[i]
    catalog.append((name, price, category))

print(catalog)


# 2. Create category to products dictionary
category_to_products = {}

for item in catalog:
    name = item[0]
    category = item[2]

    if category not in category_to_products:
        category_to_products[category] = []

    category_to_products[category].append(name)

print(category_to_products)


# 3. Category with maximum products
max_category = None
max_count = 0

for category in category_to_products:
    count = len(category_to_products[category])

    if count > max_count:
        max_count = count
        max_category = category

print(max_category)
print(category_to_products[max_category])
```
